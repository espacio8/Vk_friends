```python
import pandas as pd
import networkx as nx
import numpy as np
import csv
```

```python
df = pd.read_csv('vk_data/vk_matrix_1.csv', index_col='Name')
df.info()
df
```

```python
names = df.columns
data = np.array(df)
data.shape
```

```python
G = nx.from_numpy_matrix(data)
nx.draw(G)
```

```python
n = len(data)
closeness_centr = []
with open('vk_data/Степень близости.txt','w',encoding="utf-8") as f:  # открытие в режиме записи
    for i, w in nx.closeness_centrality(G).items():
        w_1 = round(w,4)
        s = names[i] + " " + str(w_1) + "\n"
        f.write(s)
        closeness_centr.append(w)
for idx in np.argsort(-np.array(closeness_centr))[:10]:
    print(names[idx], closeness_centr[idx])
```
