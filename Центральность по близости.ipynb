```python
import pandas as pd
import networkx as nx
import numpy as np
import csv
```

```python
df = pd.read_csv('vk_data/vk_matrix_1.csv', index_col='Name')
df.info()
df
```

```python
names = df.columns
data = np.array(df)
data.shape
```

```python
G = nx.from_numpy_matrix(data)
nx.draw(G)
```

```python
n = len(data)
max_between_centr_i = 0
between_centr = [None for i in range(n)]
f = open('vk_data/Степень близости.txt','w',encoding="utf-8")  # открытие в режиме записи
for i, w in nx.closeness_centrality(G).items():
    #print (names[i], round(w,4))
    w_1 = round(w,4)
    s = names[i] + " " + str(w_1) + "\n"
    f.write(s)
    between_centr[i] = w
    if w > between_centr[max_between_centr_i]:
        max_between_centr_i = i
f.close()
print('Максимальная степень близости:\n',names[max_between_centr_i], between_centr[max_between_centr_i])
```
