```python
import pandas as pd
import networkx as nx
import numpy as np
from collections import deque # подключаем очереди
```

```python
df = pd.read_csv('vk_data/vk_matrix_1.csv', index_col='Name')
df.info()
df
```

```python
names = df.columns
data = np.array(df)
data.shape
```

```python
names
```

```python
G = nx.from_numpy_matrix(data)
nx.draw(G)
```

```python
n = len(data)

max_degree_i = 0
degree_centr = [None for i in range(n)]

for i in range(n):
    w = data[i].sum() / (n - 1)
    degree_centr[i] = w
    if w > degree_centr[max_degree_i]:
        max_degree_i = i
    print(names[i], degree_centr[i])
```

```python
names[max_degree_i], degree_centr[max_degree_i]
```

```python
for i, w in nx.degree_centrality(G).items():
    print(names[i], round(abs(w - degree_centr[i]), 10))
```

```python
min_paths = np.full(data.shape, np.inf)
np.fill_diagonal(min_paths, 0)
min_paths
```

```python
def bfs(start, graph, min_paths):
    n = len(graph)
    queue = deque([start]) #кладем в очередь начальную вершину
    
    while queue: #пока очередь не пуста 
        i = queue.pop() #достаем первый элемент
        
        for j in range(n):
            if graph[i][j] == 1 and min_paths[start][j] == np.inf: # если у двух вершин есть связь и между ними еще не найден кратчайший путь
                min_paths[start][j] = min_paths[j][start] = min_paths[start][i] + 1
                queue.append(j)
```

```python
for i in range(n):
    if i % 10 == 0:
        print('{}/{}'.format(i + 1, n))
    bfs(i, data, min_paths)
```
