```python
import vk
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_csv('vk_data/id_names.csv', dtype=({'Id': 'int','Name': 'str'}))
```

```python
ids, names = df.Id.values, df.Name.values
```

```python
f = open('C:/Users/user/vk_token.txt', 'r')
token = f.readline().split()
```

```python
session = vk.Session(access_token = token)
vk_api = vk.API(session, v='5.87')
friends = np.zeros(shape=(len(names),len(names)))
```

```python
friends_id = []
for i, row in df.iterrows():
    try:
        friend_list = vk_api.friends.get(user_id=row[0])['items']
        friends_id.append(friend_list)
        time.sleep(0.3)
    except:
        print("Исключение было обработано")
    time.sleep(0.6)
```

```python
friends_id = [friends_part for list_in in friends_id for friends_part in list_in]
```

```python
a_id = np.array(friends_id)
ids_uniq = np.unique(a_id) #список уникальных id
```

```python
print(ids_uniq)
```

```python
row = np.arange(ids_uniq.shape[0])
print(row)
id_to = dict(zip(ids_uniq,row))
```

```python
matrix = np.zeros((ids_uniq.shape[0], ids_uniq.shape[0]))
names = []
```

```python
i = 0
for id_us, ix_us in id_to.items():
    name = vk_api.users.get(user_id=id_us)[0]
    name = name['first_name'] + " " + name['last_name']
    names.append(name)
    i+=1
    print(i/len(id_to.items())) #сколько осталось до завершения процесса
    try:
        friends_of_user = vk_api.friends.get(user_id=id_us)['items']
        user_friends_id = (np.intersect1d(ids_uniq, friends_of_user)) #ищем пересечение друзей id_us и 1875 уникальных друзей из нашего списка
        for friend_id in user_friends_id:
            friend_idx = id_to[friend_id] #ищем в словаре идентификатор друга
            matrix[friend_idx][ix_us] = 1 #заполняем матрицу смежности
            matrix[ix_us][friend_idx] = 1
    except:
        print("Исключение было обработано")
    time.sleep(0.6)

pd.DataFrame(matrix, columns=names, index=names).to_csv('vk_data/vk_matrix_1.csv', index_label='Name')
```
