# Реализация bfs

```python
min_paths = np.full(data.shape, np.inf)
np.fill_diagonal(min_paths, 0)
min_paths
```

```python
#Adj matrix to adj list
def matrix_to_list(matrix):
    graph = {}
    for i, node in enumerate(matrix):
        adj = []
        for j, connected in enumerate(node):
            if connected:
                adj.append(j)
        graph[i] = adj
    return graph
```

```python
adj_list = matrix_to_list(data)
```

```python
list(adj_list.items())[:4]
```

```python
# finds shortest path between 2 nodes of a graph using BFS
def bfs_shortest_path(graph, start, goal):
    # посещенные вершины
    explored = []
    # храним информацию о всех путях, которые нужно посетить
    queue = [[start]]
 
    # проверяем все возможные пути
    while queue:
        # берём первый путь из очереди
        path = queue.pop(0)
        # последняя вершина из пути
        node = path[-1]
        if node not in explored:
            neighbours = graph[node]
            # проходимо через все соседние вершины, создаем путь и добавлеяем его в очередь
            for neighbour in neighbours:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)
                # возвращаем путь, если соседняя вершина является целевой
                if neighbour == goal:
                    return new_path
 
            # отмечаем что посетили вершину
            explored.append(node)
    return -1 #Если пути нет
bfs_shortest_path(adj_list, 1, 2)
```

```python
paths = np.zeros((len(data),len(data)))
for i in tqdm_notebook(range(len(data))):
    for j in range(len(data)):
        if i != j:
            path = bfs_shortest_path(adj_list, i, j)
            if type(path) == list:
                paths[j, i] = len(path) - 1 
                paths[i, j] = len(path) - 1
print(paths)
```

```python
def bfs(start, graph, min_paths):
    n = len(graph)
    queue = deque([start]) #кладем в очередь начальную вершину
    
    while queue: #пока очередь не пуста 
        i = queue.pop() #достаем первый элемент
        
        for j in range(n):
            if graph[i][j] == 1 and min_paths[start][j] == np.inf: # если у двух вершин есть связь и между ними еще не найден кратчайший путь
                min_paths[start][j] = min_paths[j][start] = min_paths[start][i] + 1
                queue.append(j)
```

```python
for i in range(n):
    if i % 10 == 0:
        print('{}/{}'.format(i + 1, n))
    bfs(i, data, min_paths)
```
